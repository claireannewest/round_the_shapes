```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shape = np.loadtxt('shape.dat', skiprows=7)
slice0 = np.where(shape[:,1] == 0)[0]
x = shape[slice0,1]
y = shape[slice0,2]
z = shape[slice0,3]
rad=20
```

```python
fig, ax = plt.subplots(1,2, figsize=[12,8])
plot_circle = True

# Find center of the circle
entireh = -min(y)
h = entireh - np.sqrt(3)*rad/2 - rad / (2*np.sqrt(3))
xc = -h
yc = 0

# Find points where circle meets hexagon border 
x1 = int(xc - np.sqrt(3)*rad/2)
y1 = int(rad/2)
x2 = x1
y2 =  -y1
print(entireh, h)

### Define 'orange' points ###
idxp = np.where((y < x1) & ((y-xc)**2 + (z-yc)**2 > rad**2) )[0]

## Delete the orange points from the blue points (newpnts) ##
pnts_orange = np.column_stack((y[idxp], z[idxp]))
idx2remove = []
newpnts = shape[slice0,2:4]
for j in range(len(pnts_orange)):
    idx_remove = np.where((shape[slice0,2:4] == pnts_orange[j,:] ).all(axis=1))
    idx2remove.append(idx_remove[0][0])
newpnts = np.delete(newpnts, idx2remove, 0)

##############################################
ax[0].scatter(y, z, marker='s', s=9)
if plot_circle == True:
    ycircle = []; zcircle = []
    for i in range(len(shape[slice0,2])):
        yi = shape[slice0,2][i]
        zi = shape[slice0,3][i]
        if (yi-xc)**2 + (zi-yc)**2 < rad**2:
            ycircle.append(yi)
            zcircle.append(zi)    
    ax[0].scatter(ycircle, zcircle, color='black', marker='s', s=9, alpha=.2)
    
ax[0].scatter(xc, yc, marker='s', s=9)
ax[0].scatter(x1, y1, marker='s', s=9)
ax[0].scatter(x2, y2, marker='s',s=9)
ax[0].plot([x1,x2], [y1, y2], color='magenta')
ax[0].scatter(y[idxp], z[idxp], alpha=.3,marker='s', color='tab:orange')

ax[1].scatter(newpnts[:,0], newpnts[:,1])
ax[0].axis('equal'); ax[1].axis('equal')
ax[0].set_title('Starting');
ax[1].set_title('Trimmed');
```
